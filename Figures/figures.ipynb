# Save these as .png files!

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([[2.5, -.801],
               [3.5, -.9365],
               [4.5, -.212],
              [5.5, .71],
               [6.5, .971]])
plt.scatter(x[:,0], x[:,1])
plt.xlim(2,7)
plt.savefig('data_to_fit.png')
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,5))

x = np.array([[2.5, -.801],
               [3.5, -.9365],
               [4.5, -.212],
              [5.5, .71],
               [6.5, .971]])
for ax in fig.axes:
    ax.scatter(x[:,0], x[:,1])
    ax.set_xlim(2,7)

x1 = np.linspace(0,8,100)
y1 = np.cos(x1)
ax1.plot(x1,y1, c='g')

x2 = np.array([[2,-1.5],
              [7,1.2]])
ax2.plot(x2[:,0], x2[:,1], c='g')
```
